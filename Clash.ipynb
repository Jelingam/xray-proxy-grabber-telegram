```python
import json
from requests import get
from tqdm import tqdm
from requests.adapters import HTTPAdapter
from requests import Session
```

```python
# find all subscription link from a repo forks
forks_url = []
for page_num in tqdm(range(1,10)):
  url = f"https://api.github.com/repos/aiirobyte/TopFreeProxies/forks?per_page=100&page={page_num}"
  res = get(url)
  j = res.json()
  if len(j) == 0:
    break
  for item in j:
    forks_url.append(item["full_name"])


sub_urls = []
error = []
for u in tqdm(forks_url):
  url = f"https://raw.githubusercontent.com/{u}/master/sub/sub_list.json"
  res = get(url)
  if res.status_code == 200:
    try:
      j = res.json()
      for item in j:
        if "url" in item.keys():
          us = item["url"].split("|")
          for u in us:
            sub_urls.append(u)
    except:
      print(f"{url} can't be retrived")
      error.append(url)

sub_urls = list(set(sub_urls))
print(f"\n{len(sub_urls)} sub url found")

with open("./gdrive/MyDrive/Colab Notebooks/Clash/sub_url_all.txt", "w") as f:
  for item in sub_urls:
    f.write(item)
```

```python
with open("./gdrive/MyDrive/Colab Notebooks/Clash/sub_url_all.txt", "r") as f:
  sub_urls = f.readlines()
suc = []
err = []
for url in tqdm(sub_urls):
  try:
    res = get(url.strip(), timeout=5)
    if res.status_code == 200:
      suc.append(url)
  except:
    err.append(url)


with open("./gdrive/MyDrive/Colab Notebooks/Clash/sub_url_working.txt", "w") as f:
  for item in suc:
    f.write(item)
with open("./gdrive/MyDrive/Colab Notebooks/Clash/sub_url_notworking.txt", "w") as f:
  for item in err:
    f.write(item)
```

```python
import os
import json
from tqdm import tqdm
with open("./gdrive/MyDrive/Colab Notebooks/Clash/sub_url_working.txt", "r") as f:
  suc = f.readlines()
working_servers = []
all_servers = []
counter = 1

with open("./gdrive/MyDrive/Colab Notebooks/Clash/config.json", "r") as fr:
  con = json.load(fr)

for url in tqdm(suc):
  url = url.strip()
  if url == "":
    continue
  con["subscription"] = url
  js = json.dumps(con, indent=4)
  with open("./gdrive/MyDrive/Colab Notebooks/Clash/config.json", "w") as fw:
    fw.write(js)
  cmd = 'sudo nohup ./lite-linux-amd64 --config "./gdrive/MyDrive/Colab Notebooks/Clash/config.json" --test ss'
  os.system(cmd)
  with open("out.json", encoding="utf-8") as f:
    js = json.load(f)
  if type(js) == list:
    for item in js:
      all_servers.append(item["Link"])
      if item["ping"] != "0":
        working_servers.append(item["Link"])
  if counter % 100 == 0:
    name = f"./gdrive/MyDrive/Colab Notebooks/Clash/working_server_{counter//100}.txt"
    with open(name, "w", encoding="utf-8") as f:
      for l in working_servers:
        f.write(l)
    working_servers=[]

    name = f"./gdrive/MyDrive/Colab Notebooks/Clash/all_server_{counter//100}.txt"
    with open(name, "w", encoding="utf-8") as f:
      for l in all_servers:
        f.writelines(l)
    all_servers=[]
  counter += 1
```

```python
name = f"./gdrive/MyDrive/Colab Notebooks/Clash/working_server_{counter//100 + 1}.txt"
with open(name, "w", encoding="utf-8") as f:
  for l in working_servers:
    f.write(l)

name = f"./gdrive/MyDrive/Colab Notebooks/Clash/all_server_{counter//100 + 1}.txt"
with open(name, "w", encoding="utf-8") as f:
  for l in all_servers:
    f.writelines(l)
```
